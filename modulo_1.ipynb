**Sintaxis de la funci√≥n `drop ()` en pandas:**

`DataFrame.drop(labels=None, axis=0, index=None, columns=None, level=None, inplace=False, errors=‚Äôraise‚Äô)`

* **labels:**  Cadena o lista de cadenas que hacen referencia a la fila.
* **axis:**  int o valor de cadena, 0 '√≠ndice' para Filas y 1 'columnas' para Columnas.
* **index o columns:**  etiqueta √∫nica o lista. El √≠ndice o las columnas son una alternativa al eje y no se pueden usar juntos.
* **level:** se  utiliza para especificar el nivel, en caso de que el marco de datos tenga un √≠ndice de niveles m√∫ltiples.
* **inplace:** realiza  cambios en el marco de datos original si es verdadero.
* **errores:**  ignora el error si alg√∫n valor de la lista no existe y elimina el resto de los valores cuando `errores = 'raise'`

### **1.5.1 JOIN EN PANDAS: MERGE DATA FRAMES (INNER, OUTER, RIGHT, LEFT JOIN)**

Podemos unir o fusionar dos marcos de datos en pandas python usando la funci√≥n `merge()`. Los diferentes argumentos para `merge ()` le permiten realizar una uni√≥n natural, uni√≥n izquierda, uni√≥n derecha y uni√≥n externa completa en pandas. Tambi√©n hemos visto otros tipos de operaciones de uni√≥n o concatenaci√≥n como uni√≥n basada en √≠ndice, √≠ndice de fila e √≠ndice de columna.

### Join or Merge in Pandas ‚Äì Syntax:
`merge(left_df, right_df, on=‚ÄôCustomer_id‚Äô, how=‚Äôinner‚Äô)`

* **left_df** - Dataframe1
* **right_df** - Dataframe2.
* **on :** columnas (nombres) a las que unirse. Debe encontrarse en los objetos DataFrame izquierdo y derecho.
* **how:** el tipo de combinaci√≥n debe realizarse: 'izquierda', 'derecha', 'externa', 'interna', el valor predeterminado es la combinaci√≥n interna

Los marcos de datos deben tener los mismos nombres de columna en los que ocurre la fusi√≥n. La funci√≥n `Merge ()` en pandas es similar a la operaci√≥n de uni√≥n de base de datos en SQL.



### **1.5.2 ENTENDIENDO LOS DIFERENTES TIPOS DE JOIN OR MERGE IN PANDAS:**

* **Inner Join or Natural join:** para mantener solo las filas que coinciden con los marcos de datos, especifique el argumento `how=‚Äòinner‚Äô.`
* **Outer Join or Full outer join:** para mantener todas las filas de ambos marcos de datos, especifique `how=‚Äòouter‚Äô`.
* **Left Join or Left outer join:** para incluir todas las filas de su marco de datos xy solo las de y que coinciden, especifique `how=‚Äòleft‚Äô`.
* **Right Join or Right outer join:**para incluir todas las filas de su marco de datos y y solo las de x que coinciden, especifique `how=‚Äòright‚Äô`.
