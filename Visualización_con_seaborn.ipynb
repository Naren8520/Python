## 1- Importar librerias y conociendo los datos

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
tips = pd.read_csv("data_tip.csv")
df = tips.copy()
df.sample(7)
```

```python
df.info()
```

Visualizamos la medidas de tendencias centrales de nuestro conjunto de datos

```python
df.describe().T
```

Verificamos si tenemos valores perdido en nuestro conjunto de datos, y en caso de haber que cantidad hay para cada variable

```python
df.isnull().sum()
```

Para este conjunto de datos que estamos trabajando, se puede observar que no se tiene valores perdidos. por dicha no tenemos!!!!

```python
df.corr()
```

```python
sns.heatmap(df.corr(),annot=True)
```

## 2 -Visualization

### 2.1 Visualizando relaciones estadísticas

Usaremos la libreria `seaborn` para visualizar los diferentre gráfico haremos.

Usaremos la estilo `darkgrid` para dar una mejor apariencia a nuestro gráfico.

Usaremos la función `relplot` para visualizar la dispersión.

```python
sns.set(style="darkgrid")
sns.relplot(x = "total_bill", y = "tip", data = df);
```

En el siguiente gráfico realizamos un gráfico de dispersión para 3 variables. usamos el parámetro `hue` para agregar leyendas

```python
sns.relplot(x = "total_bill", y = "tip", hue= "smoker", data = df);
```

```python
sns.scatterplot(x = "total_bill", y = "tip", hue= "sex", data = df);
```

Podemos realizar el diagrama de dispersión también usando la función `scatterplot`

```python
sns.scatterplot(x = "total_bill", y = "tip", hue= "smoker", style= "smoker", data = df);
```

```python
sns.relplot(x = "total_bill",
            y = "tip",
            hue = "smoker",
            style = "time",
            height = 6,
            data = tips);
```

```python
sns.relplot(x = "total_bill", y = "tip", hue = "size", height = 7, data = df);
```

```python
sns.relplot(x = "total_bill", y = "tip", size = "size", sizes = (20,100), hue = "size", data = df);
```
