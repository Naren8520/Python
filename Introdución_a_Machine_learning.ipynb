### Aprendizaje Supervisado
Es la tarea de aprendizaje autom√°tico de aprender una funci√≥n que asigna una entrada a una salida en funci√≥n de pares de entrada-salida de ejemplo. En el aprendizaje supervisado, cada ejemplo es un par que consta de un objeto de entrada (t√≠picamente un vector) y un valor de salida deseado (tambi√©n llamado se√±al de supervisi√≥n). Un algoritmo de aprendizaje supervisado analiza los datos de entrenamiento y produce una funci√≥n inferida, que puede usarse para mapear nuevos ejemplos. Un escenario √≥ptimo permitir√° que el algoritmo determine correctamente las etiquetas de clase para instancias invisibles. Esto requiere que el algoritmo de aprendizaje generalice a partir de los datos de entrenamiento a situaciones invisibles de una manera "razonable". Esta calidad estad√≠stica de un algoritmo se mide mediante el llamado error de generalizaci√≥n.

#### Pasos
Para resolver un determinado problema de aprendizaje supervisado, se deben realizar los siguientes pasos:

1. Determine el tipo de ejemplos de formaci√≥n. Antes de hacer cualquier otra cosa, el debemos de decidir qu√© tipo de datos se utilizar√°n como conjunto de entrenamiento. En el caso del an√°lisis de escritura a mano, por ejemplo, esto podr√≠a ser un solo car√°cter escrito a mano, una palabra escrita a mano completa, una oraci√≥n completa de escritura a mano o quiz√°s un p√°rrafo completo de escritura a mano.

2. Definir un conjunto de entrenamiento. El conjunto de entrenamiento debe ser representativo del uso real de la funci√≥n. Por lo tanto, se recopila un conjunto de objetos de entrada y tambi√©n se recopilan las salidas correspondientes, ya sea de expertos humanos o de mediciones.

3. Determine la representaci√≥n de la caracter√≠stica de entrada de la funci√≥n aprendida. La precisi√≥n de la funci√≥n aprendida depende en gran medida de c√≥mo se representa el objeto de entrada. Normalmente, el objeto de entrada se transforma en un vector de caracter√≠sticas, que contiene una serie de caracter√≠sticas que son descriptivas del objeto. El n√∫mero de caracter√≠sticas no deber√≠a ser demasiado grande, debido a la maldici√≥n de la dimensionalidad; pero debe contener suficiente informaci√≥n para predecir con precisi√≥n la salida.

4. Determine la estructura de la funci√≥n aprendida y el algoritmo de aprendizaje correspondiente. Por ejemplo, el ingeniero puede optar por utilizar m√°quinas de vectores de soporte o √°rboles de decisi√≥n.

5. Completa el dise√±o. Ejecute el algoritmo de aprendizaje en el conjunto de entrenamiento recopilado. Algunos algoritmos de aprendizaje supervisado requieren que el usuario determine ciertos par√°metros de control. Estos par√°metros se pueden ajustar optimizando el rendimiento en un subconjunto (llamado conjunto de validaci√≥n) del conjunto de entrenamiento, o mediante validaci√≥n cruzada.

6. Eval√∫e la precisi√≥n de la funci√≥n aprendida. Despu√©s del ajuste y el aprendizaje de los par√°metros, el rendimiento de la funci√≥n resultante debe medirse en un conjunto de prueba que est√© separado del conjunto de entrenamiento.

#### Escoger el Algoritmo
Se encuentra disponible una amplia gama de algoritmos de aprendizaje supervisado, cada uno con sus fortalezas y debilidades. No existe un algoritmo de aprendizaje √∫nico que funcione mejor en todos los problemas de aprendizaje supervisado.

Hay cuatro principales problemas a considerar en el aprendizaje supervisado:

**1. Bias-variance tradeoff.** 
Un primer problema es la relaci√≥n entre sesgo y varianza. Imagine que tenemos disponibles varios conjuntos de datos de entrenamiento diferentes, pero igualmente buenos. Un algoritmo de aprendizaje est√° sesgado para una entrada particular $x$ si, cuando se entrena en cada uno de estos conjuntos de datos, es sistem√°ticamente incorrecto al predecir la salida correcta para $x$. Un algoritmo de aprendizaje tiene una alta varianza para una entrada en particular $x$ si predice diferentes valores de salida cuando se entrena en diferentes conjuntos de entrenamiento. El error de predicci√≥n de un clasificador aprendido est√° relacionado con la suma del sesgo y la varianza del algoritmo de aprendizaje. Generalmente, existe una compensaci√≥n entre sesgo y varianza. Un algoritmo de aprendizaje con un sesgo bajo debe ser "flexible" para que pueda ajustarse bien a los datos. Pero si el algoritmo de aprendizaje es demasiado flexible, se ajustar√° a cada conjunto de datos de entrenamiento de manera diferente y, por lo tanto, tendr√° una gran variaci√≥n. Un aspecto clave de muchos m√©todos de aprendizaje supervisado es que pueden ajustar esta compensaci√≥n entre sesgo y varianza.

**2. Complejidad de la funci√≥n y cantidad de datos de entrenamiento.**
El segundo problema es la cantidad de datos de entrenamiento disponibles en relaci√≥n con la complejidad de la funci√≥n "verdadera" (funci√≥n de clasificaci√≥n o regresi√≥n). Si la funci√≥n verdadera es simple, entonces un algoritmo de aprendizaje "inflexible" con alto sesgo y baja varianza podr√° aprenderlo a partir de una peque√±a cantidad de datos. Pero si la funci√≥n verdadera es muy compleja (por ejemplo, porque implica interacciones complejas entre muchas caracter√≠sticas de entrada diferentes y se comporta de manera diferente en diferentes partes del espacio de entrada), entonces la funci√≥n solo podr√° aprender de una gran cantidad de datos de entrenamiento y el uso de un algoritmo de aprendizaje "flexible" con bajo sesgo y alta varianza. Existe una clara demarcaci√≥n entre la entrada y la salida deseada.

**3. Dimensionalidad del espacio de entrada.**
Un tercer problema es la dimensionalidad del espacio de entrada. Si los vectores de caracter√≠sticas de entrada tienen una dimensi√≥n muy alta, el problema de aprendizaje puede ser dif√≠cil incluso si la funci√≥n verdadera solo depende de una peque√±a cantidad de esas caracter√≠sticas. Esto se debe a que las muchas dimensiones "adicionales" pueden confundir el algoritmo de aprendizaje y hacer que tenga una gran variaci√≥n. Por lo tanto, una dimensi√≥n de entrada alta normalmente requiere ajustar el clasificador para que tenga una baja varianza y un alto sesgo. En la pr√°ctica, si el ingeniero puede eliminar manualmente caracter√≠sticas irrelevantes de los datos de entrada, es probable que esto mejore la precisi√≥n de la funci√≥n aprendida. Adem√°s, existen muchos algoritmos para la selecci√≥n de caracter√≠sticas que buscan identificar las caracter√≠sticas relevantes y descartar las irrelevantes. Este es un ejemplo de la estrategia m√°s general de reducci√≥n de dimensionalidad, que busca mapear los datos de entrada en un espacio de menor dimensi√≥n antes de ejecutar el algoritmo de aprendizaje supervisado.

**4. Ruido en los valores de salida.**
Un cuarto problema es el grado de ruido en los valores de salida deseados (las variables objetivo de supervisi√≥n). Si los valores de salida deseados son a menudo incorrectos (debido a errores humanos o errores del sensor), entonces el algoritmo de aprendizaje no debe intentar encontrar una funci√≥n que coincida exactamente con los ejemplos de entrenamiento. Intentar ajustar los datos con demasiado cuidado conduce a un sobreajuste. Puede sobreajustar incluso cuando no hay errores de medici√≥n (ruido estoc√°stico) si la funci√≥n que est√° tratando de aprender es demasiado compleja para su modelo de aprendizaje. En tal situaci√≥n, la parte de la funci√≥n de destino que no se puede modelar "corrompe" los datos de entrenamiento; este fen√≥meno se ha denominado ruido determinista. Cuando est√° presente cualquier tipo de ruido, es mejor utilizar un estimador de mayor sesgo y menor varianza.

En la pr√°ctica, existen varios enfoques para aliviar el ruido en los valores de salida, como la parada anticipada para evitar el sobreajuste, as√≠ como detectar y eliminar los ejemplos de entrenamiento ruidosos antes de entrenar el algoritmo de aprendizaje supervisado. Hay varios algoritmos que identifican ejemplos de entrenamiento ruidosos y la eliminaci√≥n de los ejemplos de entrenamiento ruidosos sospechosos antes del entrenamiento ha disminuido el error de generalizaci√≥n con significaci√≥n estad√≠stica.

#### Otro factores a tomar encuenta sobre el Aprendizaje supervisado
Otros factores a considerar al elegir y aplicar un algoritmo de aprendizaje incluyen los siguientes:

* **Heterogeneidad de los datos.** Si los vectores de caracter√≠sticas incluyen caracter√≠sticas de muchos tipos diferentes (discretos, discretos ordenados, recuentos, valores continuos), algunos algoritmos son m√°s f√°ciles de aplicar que otros. Muchos algoritmos, incluidas las m√°quinas de vectores de soporte, la regresi√≥n lineal, la regresi√≥n log√≠stica, las redes neuronales y los m√©todos del vecino m√°s cercano, requieren que las caracter√≠sticas de entrada sean num√©ricas y escaladas a rangos similares (por ejemplo, al intervalo [-1,1]). Los m√©todos que emplean una funci√≥n de distancia, como los m√©todos del vecino m√°s cercano y las m√°quinas de vectores de soporte con n√∫cleos gaussianos, son particularmente sensibles a esto. Una ventaja de los √°rboles de decisi√≥n es que manejan f√°cilmente datos heterog√©neos.

* **Redundancia en los datos.** Si las caracter√≠sticas de entrada contienen informaci√≥n redundante (por ejemplo, caracter√≠sticas altamente correlacionadas), algunos algoritmos de aprendizaje (por ejemplo, regresi√≥n lineal, regresi√≥n log√≠stica y m√©todos basados en la distancia) funcionar√°n mal debido a inestabilidades num√©ricas. Estos problemas a menudo pueden resolverse imponiendo alguna forma de regularizaci√≥n.

* **Presencia de interacciones y no linealidades.** Si cada una de las caracter√≠sticas hace una contribuci√≥n independiente a la salida, entonces los algoritmos basados en funciones lineales (por ejemplo, regresi√≥n lineal, regresi√≥n log√≠stica, m√°quinas de vectores de soporte, Bayes ingenuo) y funciones de distancia (por ejemplo, m√©todos del vecino m√°s cercano, m√°quinas de vectores de soporte) con n√∫cleos gaussianos) generalmente funcionan bien. Sin embargo, si hay interacciones complejas entre caracter√≠sticas, los algoritmos como los √°rboles de decisi√≥n y las redes neuronales funcionan mejor, porque est√°n dise√±ados espec√≠ficamente para descubrir estas interacciones. Tambi√©n se pueden aplicar m√©todos lineales, pero el ingeniero debe especificar manualmente las interacciones al usarlos.

Al considerar una nueva aplicaci√≥n, el ingeniero puede comparar m√∫ltiples algoritmos de aprendizaje y determinar experimentalmente cu√°l funciona mejor en el problema en cuesti√≥n (ver validaci√≥n cruzada). Ajustar el rendimiento de un algoritmo de aprendizaje puede llevar mucho tiempo. Dados los recursos fijos, a menudo es mejor dedicar m√°s tiempo a recopilar datos de entrenamiento adicionales y caracter√≠sticas m√°s informativas que dedicar m√°s tiempo a ajustar los algoritmos de aprendizaje.

#### 7. C√≥mo funcionan los algoritmos de aprendizaje supervisado
Dado un conjunto de $N$ ejemplos de entrenamiento de la forma $(x_1, y_1), \cdots, (x_N, \; y_N)$ tal que $x_{i}$ es el vector de caracter√≠sticas del i-√©simo ejemplo y $y_ {i}$ es su etiqueta (es decir, clase), un algoritmo de aprendizaje busca una funci√≥n  $g: X \to Y$, donde $X$ es el espacio de entrada y $Y$ es el espacio de salida. La funci√≥n $g$ es un elemento de alg√∫n espacio de posibles funciones $G$, generalmente llamado espacio de hip√≥tesis. A veces es conveniente representar $g$ usando una funci√≥n de puntuaci√≥n $f: X \times Y \to \mathbb {R}$ tal que $g$ se define como devolver el valor $y$ que da la puntuaci√≥n m√°s alta: $g (x) = {\underset {y} {\arg \max}} \; f (x, y)$. Sea $F$ el espacio de las funciones de puntuaci√≥n.

Aunque $G$ y $F$ pueden ser cualquier espacio de funciones, muchos algoritmos de aprendizaje son modelos probabil√≠sticos donde $g$ toma la forma de un modelo de probabilidad condicional $g (x) =P (y | x)$, o $f$ toma la forma de un modelo de probabilidad conjunta $f (x, y) = P (x, y)$. Por ejemplo, Bayes ingenuo y el an√°lisis discriminante lineal son modelos de probabilidad conjunta, mientras que la regresi√≥n log√≠stica es un modelo de probabilidad condicional.

Hay dos enfoques b√°sicos para elegir $f$ o $g$: minimizaci√≥n del riesgo emp√≠rico y minimizaci√≥n del riesgo estructural. La minimizaci√≥n de riesgos emp√≠ricos busca la funci√≥n que mejor se adapta a los datos de entrenamiento. La minimizaci√≥n del riesgo estructural incluye una funci√≥n de penalizaci√≥n que controla la compensaci√≥n de sesgo / varianza.

En ambos casos, se supone que el conjunto de entrenamiento consiste en una muestra de pares independientes e id√©nticamente distribuidos, $(x_i, \; y_i)$. Para medir qu√© tan bien se ajusta una funci√≥n a los datos de entrenamiento, una funci√≥n de p√©rdida $L: Y \times Y \to \mathbb {R} ^{\geq 0}$ est√° definido. Para el ejemplo de entrenamiento $(x_i, \; y_i)$, la p√©rdida de la predicci√≥n del valor $\hat {y }$ es $L (y_i, \hat {y})$.

El riesgo $R (g)$ de la funci√≥n $g$ se define como la p√©rdida esperada de $g$. Esto se puede estimar a partir de los datos de entrenamiento como
$$R_ {emp} (g) = \frac {1} {N} \sum_i L (y_i, g (x_i)).$$

**Minimizaci√≥n de riesgos emp√≠ricos**

En la minimizaci√≥n del riesgo emp√≠rico, el algoritmo de aprendizaje supervisado busca la funci√≥n $g$ que minimiza $R (g)$. Por lo tanto, se puede construir un algoritmo de aprendizaje supervisado aplicando un algoritmo de optimizaci√≥n para encontrar $g$.

Cuando $g$ es una distribuci√≥n de probabilidad condicional $P (y | x)$ y la funci√≥n de p√©rdida es la probabilidad logar√≠tmica negativa: $L (y, \hat {y}) = - \log P (y | x)$, entonces la minimizaci√≥n del riesgo emp√≠rico es equivalente a la estimaci√≥n de m√°xima verosimilitud.

Cuando $G$ contiene muchas funciones candidatas o el conjunto de entrenamiento no es lo suficientemente grande, la minimizaci√≥n del riesgo emp√≠rico conduce a una alta varianza y una mala generalizaci√≥n. El algoritmo de aprendizaje es capaz de memorizar los ejemplos de entrenamiento sin generalizar bien. A esto se le llama sobreajuste.

**Minimizaci√≥n de riesgos estructurales**

La minimizaci√≥n del riesgo estructural busca prevenir el sobreajuste incorporando una penalizaci√≥n de regularizaci√≥n en la optimizaci√≥n. La penalizaci√≥n por regularizaci√≥n puede verse como la implementaci√≥n de una forma de navaja de Occam que prefiere funciones m√°s simples sobre las m√°s complejas.

Se ha empleado una amplia variedad de sanciones que corresponden a diferentes definiciones de complejidad. Por ejemplo, considere el caso en el que la funci√≥n $g$ es una funci√≥n lineal de la forma

$$g (x) = \sum_ {j = 1} ^ d \beta_j x_j.$$
Una penalizaci√≥n de regularizaci√≥n popular es $\sum_j \beta_j ^ 2$, que es la norma euclidiana al cuadrado de los pesos, tambi√©n conocida como $L_{2}$. Otras normas incluyen la norma $L_{1}$, $\sum_j | \beta_j |$ y $L_{0}$ "norma", que es el n√∫mero de $\beta_{j}s$ distintos de cero. La penalizaci√≥n se indicar√° con $C (g).$

El problema de optimizaci√≥n del aprendizaje supervisado es encontrar la funci√≥n $g$ que minimiza

$$J (g) = R_ {emp} (g) + \ lambda C (g).$$
El par√°metro $\lambda$ controla la compensaci√≥n de sesgo-varianza. Cuando $\lambda = 0$, esto proporciona una minimizaci√≥n del riesgo emp√≠rico con bajo sesgo y alta varianza. Cuando $\lambda$ es grande, el algoritmo de aprendizaje tendr√° un alto sesgo y una baja varianza. El valor de $\lambda$ se puede elegir emp√≠ricamente mediante validaci√≥n cruzada.

La penalizaci√≥n por complejidad tiene una interpretaci√≥n bayesiana como la probabilidad previa logar√≠tmica negativa de $g$, $- \log P (g)$, en cuyo caso $J (g)$ es la probabilidad posterior de $g$.

### Unsupervised vs. supervised vs. semi-supervised learning
El aprendizaje no supervisado y el aprendizaje supervisado con frecuencia se discuten juntos. Un bosque aleatorio es una t√©cnica de aprendizaje autom√°tico que se utiliza para resolver problemas de regresi√≥n y clasificaci√≥n. Utiliza el aprendizaje en conjunto, que es una t√©cnica que combina muchos clasificadores para proporcionar soluciones a problemas complejos.

Un algoritmo de bosque aleatorio consta de muchos √°rboles de decisi√≥n. El 'bosque' generado por el algoritmo de bosque aleatorio se entrena a trav√©s de bagging o de bootstrap. Bagging es un meta-algoritmo de conjunto que mejora la precisi√≥n de los algoritmos de aprendizaje autom√°tico.

El algoritmo (bosque aleatorio) establece el resultado en funci√≥n de las predicciones de los √°rboles de decisi√≥n. Predice tomando el promedio o la media de la producci√≥n de varios √°rboles. Aumentar el n√∫mero de √°rboles aumenta la precisi√≥n del resultado.

Un bosque aleatorio erradica las limitaciones de un algoritmo de √°rbol de decisi√≥n. Reduce el sobreajuste de los conjuntos de datos y aumenta la precisi√≥n. Genera predicciones sin requerir muchas configuraciones en paquetes (como scikit-learn).

### **¬øPor qu√© utilizar algoritmos de bosque aleatorio?**

Los algoritmos de bosque aleatorio son particularmente √∫tiles cuando:

1. Tiene muchos datos faltantes o un conjunto de datos desequilibrado.
2. Con una gran cantidad de √°rboles, puede evitar el sobreajuste durante el entrenamiento. El sobreajuste ocurre cuando los modelos de aprendizaje autom√°tico funcionan mejor en el conjunto de entrenamiento pero en el conjunto de prueba.
3. El algoritmo de bosque aleatorio se puede utilizar cuando tiene datos dimensionales muy altos.
4. A trav√©s de la validaci√≥n cruzada, el algoritmo de bosque aleatorio puede devolver mayor acc
5. El algoritmo de bosque aleatorio puede resolver tanto la clasificaci√≥n como la regresi√≥n y encuentra su aplicaci√≥n en una variedad de tareas que van desde la detecci√≥n de fraude con tarjetas de cr√©dito, la predicci√≥n del mercado de valores y la b√∫squeda de transacciones fraudulentas en l√≠nea.

### **Desventajas de los algoritmos de bosque aleatorio**

Hay dos desventajas principales de los algoritmos de bosque aleatorio:

1. El uso de una gran cantidad de √°rboles puede ralentizar el algoritmo.
2. El algoritmo de bosque aleatorio es un algoritmo predictivo, que solo puede predecir el futuro y no puede explicar lo que sucedi√≥ en el pasado utilizando el conjunto de datos.

### **Implementaci√≥n de regresi√≥n lineal con Sklearn**

Para implementar la regresi√≥n lineal con `Sklearn`, puede usar la clase LinearRegression del m√≥dulo sklearn.linear_model. Para entrenar el algoritmo, los conjuntos de entrenamiento y prueba, es decir, `X_train` y `X_test` en nuestro caso, se pasan al m√©todo `fit ()` del objeto de la clase LinearRegression. El conjunto de prueba se pasa al m√©todo `predict ()` de la clase para realizar predicciones. El proceso de entrenamiento y realizaci√≥n de predicciones con el algoritmo de regresi√≥n lineal es el siguiente:

```python
from sklearn.model_selection import train_test_split as holdout
from sklearn.ensemble import RandomForestRegressor as rfr

#Mandamos llamar al objeto

#LLamamos a la funci√≥n
Rfr = rfr(n_estimators = 5000, #El n√∫mero de √°rboles en el bosque.
          criterion = "squared_error", #reducci√≥n de varianza como criterio de selecci√≥n de caracter√≠sticas
          random_state = 2021,  #aleatoriedad utilizadas al construir √°rboles
          max_features='sqrt', #caracter√≠sticas a considerar al buscar la mejor divisi√≥n
          n_jobs = -1) #significa utilizar todos los procesadores

#Entrenamos el modelo
model_rfr=Rfr.fit(x_train,y_train)

#Hacemos las prediciones del modelo
y_pred_rfr = model_rfr.predict(x_test)
```

```python
print('Feature importance ranking\n\n')
importances = Rfr.feature_importances_
std = np.std([tree.feature_importances_ for tree in Rfr.estimators_],axis=0)
indices = np.argsort(importances)[::-1]
variables = ['age', 'sex', 'bmi', 'children','smoker', 'region']
importance_list = []
for f in range(x.shape[1]):
    variable = variables[indices[f]]
    importance_list.append(variable)
    print("%d.%s(%f)" % (f + 1, variable, importances[indices[f]]))
 
# Plot the feature importances of the forest
plt.figure(figsize=(15,8))
plt.title("Variables m√°s importantes")
plt.bar(importance_list, importances[indices],
       color="y", yerr=std[indices], align="center")
```

### **Tipos de Metricas**

Una vez que haya entrenado un modelo y haya hecho predicciones en el conjunto de prueba, el siguiente paso es saber qu√© tan bien se ha desempe√±ado su modelo para hacer predicciones en el conjunto de prueba desconocido. Hay varias m√©tricas para comprobarlo. Sin embargo, el error absoluto medio, el error cuadr√°tico medio y el error cuadr√°tico medio medio son tres de las m√©tricas m√°s comunes.

**Error absoluto medio(Mean Absolute Error MAE)**

El error absoluto medio (MAE) se calcula tomando el promedio del error absoluto obtenido al restar los valores reales de los valores predichos. La ecuaci√≥n para calcular MAE es:

$$MAE=\frac{\Sigma_{i=1}^{n} |y_i-\hat{y}|}{n} $$

**Error medio cuadrado (Mean Squared Error MSE)**

El error cuadr√°tico medio (MSE) es similar al MAE. Sin embargo, el error para cada registro se eleva al cuadrado en el caso de MSE para castigar los registros de datos con una gran diferencia entre los valores predichos y reales. La ecuaci√≥n para calcular el error cuadr√°tico medio es la siguiente:

$$MSE=\frac{1}{2} \Sigma_{i=1}^{n} (y_i-\hat{y})^2 $$

**Root Mean Squared Error**

El error cuadr√°tico medio es simplemente la ra√≠z inferior del error cuadr√°tico medio y se puede calcular de la siguiente manera:

$$RMSE=\sqrt{\frac{1}{n} \Sigma_{i=1}^{n} (y_i-\hat{y})^2  } $$

Los m√©todos utilizados para encontrar el valor de estas m√©tricas est√°n disponibles en la clase `sklearn.metrics`. Los valores predichos y reales deben pasarse a estos m√©todos, como se muestra en el resultado.

```python
from sklearn import metrics
from sklearn.metrics import mean_squared_error, r2_score

print ("Mean Absolute Error:" , metrics.mean_absolute_error(y_test, y_pred_rfr))
print ("Mean Squared Error:" , metrics.mean_squared_error(y_test, y_pred_rfr))
print ("Root Mean Squared Error:" , np.sqrt(metrics.mean_squared_error(y_test, y_pred_rfr)))
# The coefficient of determination: 1 is perfect prediction
print('Coefficient of determination: %.2f'% r2_score(y_test, y_pred_rfr))
#print("Mean absolute percentage error:", mean_absolute_percentage_error(y_test, y_pred_lineal) )
```
